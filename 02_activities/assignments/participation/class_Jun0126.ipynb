```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
# make some sample data
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0, 100, 50)
```

```python
#basic scatter plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x, y)
```

```python
#modifying graph (bar plot)
fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(x, y)
```

```python
#modifying graph (line plot)
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y)
ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
fig.tight_layout()
```

```python
#modifying graph (histogram
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y)
```

```python
#font dictionatires
font1 = {'family': 'self-serif', 'color': 'blue', 'size': 20}
font2 = {'family': 'monospace', 'color': 'green', 'size': 14}
```

```python
#modifying graph (line plot) with font dictionaries
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x, y)
ax.set_title('Total growth over time', fontdict=font1, loc='center')
ax.set_ylabel('Total growth', fontdict=font2)
ax.set_xlabel('Years since start', fontdict=font2)
fig.tight_layout()
```

```python
#customizing plot appearance
fig, ax = plt.subplots(figsize=(5, 3))

ax.scatter(x, y,
           marker='*',
           color='indigo')
fig.show()
```

```python
#customizing plot appearance(line plot)
fig, ax = plt.subplots(figsize=(5, 3))

ax.plot(x, y,
        marker='*',
        color='lime',
        linestyle='--', #new additions
        linewidth=2)    #new additions
fig.show()
```

```python
#customizing plot appearance(line plot) color change
fig, ax = plt.subplots(figsize=(5, 3))

ax.plot(x, y,
        marker='*',
        markersize=12,
        color="#c4165f",
        linestyle='--',
        linewidth=2,
        markeredgecolor='#fa9359',
        markerfacecolor='#000000')
fig.show()
```

```python
#grid lines
fig,ax=plt.subplots(figsize=(5,3))
ax.grid(axis = 'y', color='blue', linewidth=2, linestyle='-.')
ax.plot(x, y,
        marker='*',
        markersize=12,
        color="#c4165f",
        linestyle='--',
        linewidth=2,
        markeredgecolor='#fa9359',
        markerfacecolor='#000000')
```
