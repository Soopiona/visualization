```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
tips = sns.load_dataset('tips')
print(tips)
sns.set_style('dark')
tipgraph=sns.lineplot(data=tips, x='total_bill', y='tip',
                      color = 'hotpink',
                      linestyle = '--',
                      linewidth = 3,
                      marker = 'o',
                      markerfacecolor = 'indigo')
tipgraph.set(title='Total Bill vs Tip', xlabel='Total Bill ($)', ylabel='Tip amount($)')
plt.show()
```

```python
tipgraph=sns.scatterplot(data=tips, x='total_bill', y='tip',
                         style='time', hue='day',
                         palette=['purple','hotpink','deepskyblue','yellowgreen'])
tipgraph.set(title='Total Bill vs Tip', xlabel='Total Bill ($)', ylabel='Tip amount($)')
```

```python
sns.pairplot(data=tips, hue='day')
```

```python
daysplot=sns.relplot(data=tips, x='total_bill', y='tip',
                     hue='sex', col='day', kind='scatter', col_wrap=2)
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests


import plotly.graph_objects as go
from wordcloud import WordCloud
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted
```

```python
x1 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])

graph=go.Figure()
graph.add_trace(go.Bar(x=x1, y=y1))

graph.update_layout(title='Pirate Scores',
                    xaxis_title='Pirates',
                    yaxis_title='Score')

graph.write_html("pirategraph.html")
```

```python
graph=go.Figure()
graph.add_trace(go.Scatter(x=x1, y=y1, mode='markers'))

graph.update_layout(title='Interactive Pirate Plot',
                    xaxis_title='Pirates',
                    yaxis_title='Score',
                    width=500,height=500)
```
