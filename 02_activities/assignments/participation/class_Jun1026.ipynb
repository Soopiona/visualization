Customizing our plots with the previous lesson scripts

```python
fig,(ax1,ax2) = plt.subplots(ncols=2, nrows=1, figsize=(7,3))

ax1.scatter(x1, y1,
            marker='*',
            color='indigo')
ax2.bar(x2, y2, color='pink')
ax2.set_title('Bar Plot')
fig.show()
```

```python
fig,(ax1,ax2) = plt.subplots(ncols=2, nrows=1, figsize=(7,3))
ax1.scatter(x1,y1)
ax1.annotate("This is important", xy=(10,60), xytext=(20,60), arrowprops=dict(facecolor='black'))
ax2.bar(x2,y2)
fig.show()            
```

```python
fig, someaxes=plt.subplot_mosaic([['ax1','ax3'], ['ax2','ax3']], figsize=(7,4), layout='constrained')

someaxes['ax1'].scatter(x1,y1)
someaxes['ax2'].bar(x2,y2)
someaxes['ax3'].plot(x1,y1)
someaxes['ax1'].set_xlabel('A Big Lable', fontsize=18)
someaxes['ax2'].set_xlabel('Another Label', fontsize=18)
someaxes['ax3'].set_xlabel('Lable 2: 2 Fast 2 Furious', fontsize=18)
```

```python
#first make our sample data
x=np.array(["Luffy","Zoro","Nami","Usopp","Sanji"])
y1=np.array([110,180,240,99,220])
y2=np.array([170,100,90,120,50])

#define our figure and axes (just one this time)
fig, ax = plt.subplots(figsize=(7, 3))

#now call both bar and plot elements to the same axes (ax)
ax.bar(x, y1, color='indigo')
ax.plot(x, y2, color='orange')
```

Error imformation_errorbar

```python
y2_sd=np.std(y2)
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color='orange')
```

```python
y2_sd=np.std(y2)
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color='orange')
ax.errorbar(x, y2, 
            yerr=y2_sd, 
            fmt='o', 
            ecolor='red', 
            elinewidth=2, 
            capsize=3, 
            capthick=3)
```

Errorevery

```python
y2_sd=np.std(y2)
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color='orange')
ax.errorbar(x, y2, 
            yerr=y2_sd, 
            fmt='o', 
            ecolor='red', 
            elinewidth=2, 
            capsize=3, 
            capthick=3,
            errorevery=2)
```

Image from the internet

```python
from PIL import Image
import requests
from io import BytesIO

url = "https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png"

headers = {
    "User-Agent": "Mozilla/5.0",
    "Accept": "image/*,*/*;q=0.8",
    "Referer": "https://en.wikipedia.org/"
}

response=requests.get(url,headers=headers)
image_file=BytesIO(response.content)
image=Image.open(image_file)
image
```
